# **Exlanatory Data Analysis**

# **Gender**

```python
df.columns
```

```python
# Create the count plot
ax = sns.countplot(x='Gender', data=df)

# Define custom colors (one per bar)
colors = ['skyblue', 'lightgreen']  

# Apply the colors to each bar
for bar, color in zip(ax.patches, colors): 
    bar.set_facecolor(color)
    
# Label the bars
for bars in ax.containers:
    ax.bar_label(bars)
#Title 
plt.title('Total Sales by Gender')
plt.show()
```

```python
# Grouping and summing
Sales_gen = df.groupby('Gender', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plot the barplot and get the Axes object
ax = sns.barplot(x='Gender', y='Amount', data=Sales_gen)

# Define custom colors (one per bar)
colors = ['skyblue', 'pink']  

# Apply the colors to each bar in the Axes object
for bar, color in zip(ax.patches, colors):
    bar.set_facecolor(color)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)
# Add title
ax.set_title('Total Sales by Gender')
plt.show()
```

#### From above Graphs we can see that most of the buyers are female and the purchasing of females is greater than men

## Age Group

```python
df.columns
```

```python
# Create the countplot for AgeGroup
ax = sns.countplot(x='Age Group', data=df)

# Define custom colors — one per age group
colors = ['skyblue', 'lightgreen', 'salmon', 'purple', 'orange']  # length should match unique age groups

# Apply the colors to each bar
for bar, color in zip(ax.patches, colors):
    bar.set_facecolor(color)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)

plt.title('Age Group Distribution')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Group and sum Amount by AgeGroup
Sales_age = df.groupby('Age Group', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Plot barplot and get Axes
ax = sns.barplot(x='Age Group', y='Amount', data=Sales_age)

# Define colors — make sure to match number of age groups
colors = ['skyblue', 'lightgreen', 'salmon', 'purple', 'orange']

# Apply colors to bars
for bar, color in zip(ax.patches, colors):
    bar.set_facecolor(color)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)    

# Add title and labels
ax.set_title('Total Sales by Age Group')
ax.set_xlabel('Age Group')
ax.set_ylabel('Total Sales Amount')

plt.show()
```

#### From above graph we cn see that most of the buyers are from age group 26-35 years female

# State

```python
sns.set(style='whitegrid', rc={'figure.figsize': (15, 5)})

orders_per_state = df['State'].value_counts().nlargest(10).reset_index()
orders_per_state.columns = ['State', 'Order_Count']

palette = sns.color_palette('Set2', n_colors=10)

ax = sns.barplot(
    x='State',
    y='Order_Count',
    data=orders_per_state,
    hue='State',         # fix warning by assigning hue
    palette=palette,
    dodge=False,
    legend=False
)
# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)

ax.set_title('Total Number of Orders from Top 10 States')
ax.set_xlabel('State')
ax.set_ylabel('Number of Orders')

plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```

```python
# Set style and figure size
sns.set(style='whitegrid', rc={'figure.figsize': (15, 5)})

# Group by State and sum Amount, then get top 10 states by sales amount
sales_per_state = (
    df.groupby('State', as_index=False)['Amount']
      .sum()
      .sort_values(by='Amount', ascending=False)
      .head(10)
)

# Define color palette
palette = sns.color_palette('Set2', n_colors=10)

# Plot with hue='State' to avoid warning and legend=False to hide legend
ax = sns.barplot(
    x='State',
    y='Amount',
    data=sales_per_state,
    hue='State',
    palette=palette,
    dodge=False,
    legend=False
)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)


ax.set_title('Total Sales Amount from Top 10 States')
ax.set_xlabel('State')
ax.set_ylabel('Total Sales Amount')

plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()

```

#### From above graph we can see that most of the sales and total sales/Amount are from Uttar Pradesh, Maharashtra, Karnataka.

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Set style and figure size
sns.set(style='whitegrid', rc={'figure.figsize': (20, 5)})

# Define color palette
palette = sns.color_palette('Set2', n_colors=10)

# Create the countplot for Marital_Status with the palette
ax = sns.countplot(x='Occupation', data=df, palette=palette)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)


# Set x-axis label
plt.xlabel('Occupation')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Group and sum Amount by Occupation
Sales_Occupation = df.groupby('Occupation', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

# Set style and figure size
sns.set(style='whitegrid', rc={'figure.figsize': (20, 5)})

# Define a color palette with enough colors
palette = sns.color_palette('husl', n_colors=Sales_Occupation['Occupation'].nunique())

# Plot barplot with hue set to Occupation (and hide the redundant legend)
ax = sns.barplot(
    x='Occupation',
    y='Amount',
    hue='Occupation',
    data=Sales_Occupation,
    palette=palette,
    legend=False  # No need to show a legend since Occupation is already on x-axis
)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)


# Add title and labels
ax.set_title('Total Sales by Occupation')
ax.set_xlabel('Occupation')
ax.set_ylabel('Total Sales Amount')

# Show the plot
plt.show()
```

#### From above graph we can see that most of the buyers are from ITSector, Healthcare, Aviation

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Optionally wrap long labels for better spacing
df['Product_Category'] = df['Product_Category'].apply(lambda x: '\n'.join(x.split()))

# Set style and figure size
sns.set(style='whitegrid', rc={'figure.figsize': (20, 5)})

# Define color palette
palette = sns.color_palette('Set2', n_colors=10)

# Create the countplot
ax = sns.countplot(x='Product_Category', data=df, palette=palette)

# Set x-axis label
plt.xlabel('Product_Category')

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2., height + 10,  # slight padding above the bar
            f'{int(height)}', ha='center', va='bottom', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group and sum Amount by Product Category
Sales_Product = df.groupby('Product_Category', as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

# Optional: wrap long labels
Sales_Product['Product_Category'] = Sales_Product['Product_Category'].apply(lambda x: '\n'.join(x.split()))

# Set style and figure size
sns.set(style='whitegrid', rc={'figure.figsize': (22, 5)})

# Define a color palette
palette = sns.color_palette('husl', n_colors=Sales_Product['Product_Category'].nunique())

# Plot barplot (no hue needed)
ax = sns.barplot(
    x='Product_Category',
    y='Amount',
    hue='Product_Category',    
    data=Sales_Product,
    palette=palette,
    legend=False                
)


# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2.,
        height + 10,
        f'{int(height)}',
        ha='center',
        va='bottom',
        fontsize=12
    )

# Add title and labels
ax.set_title('Total Sales by Product Category', fontsize=18)
ax.set_xlabel('Product Category')
ax.set_ylabel('Total Sales Amount')

# Show the plot
plt.tight_layout()
plt.show()
```

#### From above graph we can see that most of the buyers are from ITSector, Healthcare, Aviation

```python
# Top 10 most products sold
Sales_Orders = df.groupby('Product_ID', as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set style and figure size
sns.set(style='whitegrid', rc={'figure.figsize': (20, 5)})

# Define a color palette based on number of unique products
palette = sns.color_palette('husl', n_colors=Sales_Orders['Product_ID'].nunique())

# Plot barplot with hue set to Product_ID to use palette, hide legend
ax = sns.barplot(
    x='Product_ID',
    y='Orders',
    hue='Product_ID',
    data=Sales_Orders,
    palette=palette,
    legend=False
)

# Add values on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(
        p.get_x() + p.get_width() / 2.,
        height + 10,  # padding above bar
        f'{int(height)}',
        ha='center',
        va='bottom',
        fontsize=14
    )

# Add title and labels
ax.set_title('Total Orders by Product')
ax.set_xlabel('Product ID')
ax.set_ylabel('Total Orders')

plt.tight_layout()
plt.show()
```
